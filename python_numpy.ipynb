# NumPy

### Importação e Carregamento de Dados:

- `import numpy as np:` importa a bibloteca Numpy e usa o alias de 'np';

- `np.loadtxt(url, delimiter = , usecols= ):` carrega os dados presentes na url e gera um array da NumPy;
    - `delimiter=`: indica qual o separador dos dados;
    - `usecols=`: indica quais as colunas númericas a serem lidas;
    
- `np.savetxt('nome', var, delimiter= ):` gera um arquivo a partir dos dados que estão na variável (var);

### Arrays:

- Arrays são diferentes de listas;
- Não aceitam valores do tipo str (string);
- São ideiais para operações matemáticas;

### Funções e Métodos:

#### Métodos:

- `.ndim:` retorna o número de dimensões do array;

- `.size:` retorna a quantidade de elementos do array;

- `.shape:` retorna a quantidade de linhas e colunas do array;

- `.T:` faz a transposição de linhas e colunas do array;

- `.skiprows:` pula as linhas indicadas;


#### Funções:

Seleção:

- `nome_array[x:y, w:z]:` seleciona o intervalo de linhas e/ou colunas de um array;

<br>

Geração:

- `np.arange(inicio, fim (excl), salto):` gera um array de números inteiros;

- `np.array([]):` gera um array vazio;

- `np.copy():` gera uma cópia de um array;

- `np.set_printoptions(suppress= , precision= )`: ajusta a forma do NumPy de imprimir os números:
    - `suppress=True`: evita notação científica;
    - `precision=n`: imprime números com N casas decimais;

<br>

Comparação:

- `np.array_equal(array 1, array 2):` compara se dois arrays são iguais;

- `np.allclose(array 1, array 2, n):` compara se dois arrays são próximos de um valor N;

<br>

Valores Nulos:

- `np.isnan():` retorna True ou False em cada elemento se for NaN (valor nulo);

<br>

Operações Matemáticas e Estatísticas:

- `np.sum(), np.mean(), np.power(), np.sqrt():` realiza operações de: soma, média, potência e raiz quadrada (respectivamente);

- `np.min() e np.argmin():` retorna o valor mínimo e a posição do valor mínimo;

- `np.max() e np.argmax():` retorna o valor máximo e a posição do valor máximo;

- `np.linalg.norm():` calcula a norma entre dois arrays;

<br>

Manipulação:

- `np.append(nome_array, valor):` adiciona novos valores ao final de um array, mas gera um array novo;
    - sem axis: gera um array unidimensional (1D);
    - axis = 0: adiciona linha;
    - axis = 1: adiciona coluna;

- `np.column_stack():` une vários arrays unidimensionais (1D);

<br>

Geração de Valores Aleatórios:

- `np.random.randint(low= , high= , size= ):` gera números inteiros aleatórios dentro do intervalo indicado;

- `np.random.uniform(low= , high= , size= ):` gera números float aleatórios dentro do intervalo indicado;

- `np.random.seed():` define uma semente para geração de números aleatórios;

### Exercício 02:

Agrupando diametros e pesos para Laranja e Toranja:

```python
diametros_laranja = dados[:5000,0]
pesos_laranja = dados[:5000,1]

diametros_toranja = dados[5000:,0]
pesos_toranja = dados[5000:,1]
```

### Exercício 03:

Exemplo da função np.copy():

```python
preco_imoveis = np.array([10000,120000,11000,200000])
preco_imoveis_sao_paulo = np.copy(preco_imoveis)
print(preco_imoveis_sao_paulo)

# Mostrando que a alteração no array original, não altera a cópia:
preco_imoveis[0] = 120000
print(preco_imoveis)
print(preco_imoveis_sao_paulo)
```

Exemplo do gráfico circular:

```python
x = np.arange(-1, 1.01, 0.0001)
y1 = np.sqrt(1 - x**2)
y2 = -1 * np.sqrt(1 - x**2)
```

```python
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r' )
plt.title("Círculo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.show()
```

Regressão Linear (para TORANJA é só trocar as variaveis):

```python
Y = pesos_laranja
X = diametros_laranja
n = np.size(X)
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
b = np.mean(Y) - a*np.mean(X)
y = a * X + b
```

```python
plt.plot(diametros_laranja, pesos_laranja)
plt.plot(X, y)
plt.show()
```

Regressão Linear com Numeros Aleatorios (para TORANJA é só trocar as variaveis):

```python
# Gerando coeficientes angulares e normas:
np.random.seed(84)

coef_angulares_laranja = np.random.uniform(low=0, high=30.0, size=100)

norma_laranja = np.array([])

for i in range(100):
    norma_laranja = np.append(norma_laranja, np.linalg.norm(pesos_laranja - (coef_angulares_laranja[i]*diametros_laranja+17)))

print(norma_laranja)
print(coef_angulares_laranja)
```

```python
# Norma mínimo e melhor Coeficiente:
valor_min_laranja = norma_laranja.min()
print(format(valor_min_laranja, ".2f"))
indice_valor_laranja = norma_laranja.argmin()
print(format(coef_angulares_laranja[indice_valor_laranja], '.2f'))
```

### 10) Diferença entre arrays (regressao linear por tentativa):

Primeira Tentativa:

```python
# Primeira tentativa de função linear:
x = datas_adap
y = 2 * x + 80
```

```python
# Melhorando os números:
potencia = np.power(Moscow - y, 2)
print(potencia)

sum = np.sum(np.power(Moscow - y, 2))
print(sum)

raiz = np.sqrt(np.sum(np.power(Moscow - y, 2)))
print(raiz)
```

```python
# Verificando se a reta se aproxima do gráfico de Vendas:
plt.plot(datas_adap, Moscow)
plt.plot(x, y)
plt.show()
```

Segunda Tentativa:

```python
# Segunda tentativa:
x = datas_adap
y = 0.52 * x + 80

# Melhorando os números:
raiz = np.linalg.norm(Moscow - y)
print(raiz)
```

```python
# Verificando se a reta se aproxima do gráfico de Vendas:
plt.plot(datas_adap, Moscow)
plt.plot(x, y)
plt.show()
```

### 11) Diferença entre arrays (regressão linear por formula):

```python
# Dados iniciais:
y = Moscow
x = datas_adap
n = np.size(Moscow)
x_2 = x ** 2
```

```python
# Aplicação na fórmula:
a = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) / (n*np.sum(x_2) - np.sum(x)**2)
b = np.mean(y) - a * np.mean(x)
y = a * x + b
print(np.linalg.norm(Moscow - y))
```

```python
# Gerando o gráfico:
plt.plot(datas_adap, Moscow)
plt.plot(x, y)

# Estimando valores intermediarios e futuros:
plt.plot(41.5, 41.5*a+b, '*r')
plt.plot(100, 100*a+b, '*r')

plt.show()
```

### 12) Usando numeros aleatorios para definir a regressão linear:

Gerando coeficientes angulares e calculando a norma deles:

```python
# Gerando coeficientes angulares aleatórios:
coef_angulares = np.random.uniform(low=0.10, high=0.91, size=100)

# Array para armazenar as normas obtidas com esses coeficientes:
norma2 = np.array([])

# Preenchendo o array:
for i in range(100):
    norma2 = np.append(norma2, np.linalg.norm(Moscow - (coef_angulares[i]*x+b)))

norma2
```

Usando a norma mínima para identificar o melhor coeficiente angular:

```python
# Valor Mínimo de Norma:
min = norma2.min()
argmin = norma2.argmin()
float(coef_angulares[argmin])
```

### 13) Testando a Reprodutibilidade (sementes):

Gerando uma semente e um array de valores aleatórios de coefiente angular:

```python
np.random.seed(84)
coef_ang = np.random.uniform(low=0.10, high=0.90, size=100)
normas = np.array([])

for i in range(100):
    normas = np.append(normas, np.linalg.norm(Moscow - (coef_ang[i]*x+b)))

print(normas)
print(coef_ang)
```

Usando a norma mínima para identificar o melhor coeficiente angular:

```python
# Valor Mínimo de Norma:
min = normas.min()
argmin = normas.argmin()
float(coef_ang[argmin])
```

### 14) Salvando os resultados:

Juntando as normas e coeficientes em um único array:

```python
dados_normas_coef = np.column_stack([normas, coef_ang])
print(dados_normas_coef.shape)
np.set_printoptions(suppress=True, precision=6)
print(dados_normas_coef)
```

Salvando:

```python
np.savetxt('dados_regressao.csv', dados_normas_coef, delimiter=',')
```
